```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
a = pd.read_csv("kn.csv")


print(a.shape)
print(a['bedroom'].describe())
```

```python
c = a.corr()
print(c)
```

```python
x = a['bedroom']
y = a['price']
print(x)
print(y)

x = np.array(x).reshape(-1,1)
print(x)
```

```python
o = LinearRegression()
lm = o.fit(x, y)

print(lm)
print(lm.intercept_)
print(lm.coef_)
```

```python
b = lm.predict(x)
print(b)
```
